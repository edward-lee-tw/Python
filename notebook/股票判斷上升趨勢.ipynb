```python
from pandas_datareader import data as web
import datetime as dt
#import fix_yahoo_finance as yf
import yfinance as yf


import warnings
warnings.filterwarnings("ignore")

yf.pdr_override()

#start = dt.datetime(2020, 1, 1)
start = dt.datetime(2021, 1, 1)
#end = dt.datetime(2016, 12, 31)

stock = ['1101.TW','2303.TW']
#stock = []
#for i in range(1101,2002) :
#    j = str(i) + ".TW"
#    stock.append(j)

for k in stock :
    try : 
        # df = web.get_data_yahoo([k],start, end)
        df = web.get_data_yahoo([k],start)        
        if df.empty == False :
            # df.to_csv(k + '.csv')
            #取出data frame其中一個欄位:關盤價
            P = df["Close"]
            
            #移動平均:透過window=20參數可以將前20天的累計值計算出來
            # A=P.rolling(window=20).mean()
            # print(type(A)) #<class 'pandas.core.series.Series'>
            # print(A)

            #移動平均:透過window=20參數可以將前20天的累計值計算出來            
            df["ma_20"]=P.rolling(window=20).mean()

            #移動平均:透過window=60參數可以將前60天的累計值計算出來            
            df["ma_60"]=P.rolling(window=60).mean()

            #移動平均:透過window=120參數可以將前120天的累計值計算出來            
            df["ma_120"]=P.rolling(window=120).mean()
            
            
            # 寫檔
            df.to_csv( 'D:\\temp\\' + k + '.csv')
        else :
            pass
    except : 
        pass
```

```python
df_1=df_tickers.iloc[:,[0,1,2,5,6]]  # 整理股票的總資訊
```

```python
from pandas_datareader import data as web
import datetime as dt
# import fix_yahoo_finance as yf
import yfinance as yf
import pandas as pd

import warnings

warnings.filterwarnings("ignore")

yf.pdr_override()

# start = dt.datetime(2020, 1, 1)
start = dt.datetime(2021, 1, 1)
# end = dt.datetime(2016, 12, 31)

#stock = ['1101.TW', '2303.TW']
stock = []
for i in df_1[1] :
    j = str(i) + ".TW"
    stock.append(j)


# df_Total = pd.DataFrame({'股票代號': [], 'ma_60': []})
# df_Total.index.name='股票代號'
# Print the DataFrame
df_Total = pd.DataFrame()
df_Total['股票代號']=[]
df_Total['ma_60']=[]
print(df_Total)

index = 0
for k in stock:
    try:
        # df = web.get_data_yahoo([k],start, end)
        df = web.get_data_yahoo([k], start)
        if df.empty == False:
            # df.to_csv(k + '.csv')
            # 取出data frame其中一個欄位:關盤價
            P = df["Close"]

            # 移動平均:透過window=20參數可以將前20天的累計值計算出來
            # A=P.rolling(window=20).mean()
            # print(type(A)) #<class 'pandas.core.series.Series'>
            # print(A)

            # 移動平均:透過window=20參數可以將前20天的累計值計算出來
            df["ma_20"] = P.rolling(window=20).mean()

            # 移動平均:透過window=60參數可以將前60天的累計值計算出來
            df["ma_60"] = P.rolling(window=60).mean()

            # 60移動平均上升
            if P.rolling(window=60).mean().iloc[-2] < P.rolling(window=60).mean().iloc[-1]:
                df_Total.loc[index, "股票代號"] = k
                df_Total.loc[index, "ma_60"] = P.rolling(window=60).mean().iloc[-1]
                index+=1

            # 移動平均:透過window=120參數可以將前120天的累計值計算出來
            df["ma_120"] = P.rolling(window=120).mean()

            # 寫檔
            # df.to_csv('D:\\temp\\' + k + '.csv')
        else:
            pass
    except:
        pass

print(df_Total)
df_Total.to_csv('D:\\temp\\' +  '股票_Total.csv',encoding='big5')
```
