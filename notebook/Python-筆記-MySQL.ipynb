#### Insert Into Table

```python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="Ewl12345~",
  database="test"
)

mycursor = mydb.cursor()

sql = "INSERT INTO new_table (no, student_name, student_class, student_grade) VALUES (%s, %s, %s, %s)"
val = ("00001", "李文賜", "中文","98")
mycursor.execute(sql, val)

mydb.commit()

print(mycursor.rowcount, "record inserted.")
```

```python
sql = "INSERT INTO new_table (no, student_name, student_class, student_grade) VALUES (%s, %s, %s, %s)"
val = ("00004", "李文賜", "社會","60")
mycursor.execute(sql, val)

sql = "INSERT INTO new_table (no, student_name, student_class, student_grade) VALUES (%s, %s, %s, %s)"
val = ("00005", "李文賜", "科學","96")
mycursor.execute(sql, val)


mydb.commit()
```

#### Select From a Table

```python
mycursor.execute("SELECT * FROM new_table")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

#### Delete Record

```python
sql = "DELETE FROM new_table WHERE no = '00004'"

mycursor.execute(sql)

mydb.commit()

print(mycursor.rowcount, "record(s) deleted")
```

```python
mycursor.execute("SELECT * FROM new_table")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```

#### Update Table

```python
sql = "UPDATE new_table SET student_grade = '90' WHERE student_class = '數學'"

mycursor.execute(sql)

mydb.commit()

print(mycursor.rowcount, "record(s) affected")
```

```python
mycursor.execute("SELECT * FROM new_table")

myresult = mycursor.fetchall()

for x in myresult:
  print(x)
```
