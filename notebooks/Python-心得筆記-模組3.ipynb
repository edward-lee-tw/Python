```python
def isPhoneNumber(text):
    if len(text) != 12:
        return False
    for i in range(0,2):
        if not text[i].isdecimal():
            return False
    if text[2] != '-':
        return False
    for i in range(3,7):
        if not text[i].isdecimal():
            return False
    if text[7] != '-':
        return False
    for i in range(8,12):
        if not text[i].isdecimal():
            return False
    return True 
 
print('this is a number: 02-1234-5555')
print(isPhoneNumber('02-1234-5555'))
print('this is a number: oh no! regular expression')
print(isPhoneNumber('oh no! regular expression'))
```

```python
import re
phoneNumRegex = re.compile(r'\d\d-\d\d\d\d-\d\d\d\d')   # 樣板:r'\d\d-\d\d\d\d-\d\d\d\d'
#print(type(phoneNumRegex))  #<class 're.Pattern'>
result = phoneNumRegex.findall('Please call David at 02-8888-1688 by today. 02-9888-9898 is his office number.')
#print(type(result))  #<class 'list'>
number = ', '.join(result)
print('Phone Number: ' + number)
print('Please call him/her as soon as possible.')
```

```python
import re
def is_number(num):
  pattern = re.compile(r'(.*)\.(.*)\.(.*)\.(.*)')
  if num.count('.') > 1:
    return False
  elif pattern.match(num):
    return False
  return num.replace(".","").isdigit()

print(is_number('0.2'))
print(is_number('2.p'))
print(is_number('0.2.2'))
print(is_number('+0.2.2'))
```

```python
import re
phoneNumRegex = re.compile(r'^\d*\.?\d+$')   # 樣板: 數字
#print(type(phoneNumRegex))  #<class 're.Pattern'>
result = phoneNumRegex.findall('23.')
#print(type(result))  #<class 'list'>
print(phoneNumRegex.findall('23.'))
print(phoneNumRegex.findall('.23'))
print(phoneNumRegex.findall('7.abc'))
print(phoneNumRegex.findall('鴻海100.0台積410.3中信25.6'))
print(phoneNumRegex.findall('鴻海 100.0 台積 410.3 中信 25.6'))
```
