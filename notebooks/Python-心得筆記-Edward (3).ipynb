# 長表格? 寬表格?
### 寬表格是一種資料視覺化(報表)，而長表格是做為清理乾淨的原始資料，則兩者之間就存在轉換的關係

[請參考](https://daxpowerbi.com/%E9%95%B7%E8%A1%A8%E6%A0%BC-%E5%AF%AC%E8%A1%A8%E6%A0%BC-%E5%A5%BD%E7%9A%84%E8%A1%A8%E6%A0%BC%E8%AE%93%E4%BD%A0%E7%8E%A9%E8%BD%89%E8%B3%87%E6%96%99%E7%A7%91%E5%AD%B8%EF%BC%81/)

### 寬表格是資料視覺化的一種報表，因此寬表格的特性是每一列(row)包含多個觀察值
### 長表格如上圖是乾淨的資料(tidy data)，其特性是每一列(row)只包含一個觀察值，根據RStudio的首席資料科學家Hadley Wickham在Journal of Statistical Software中所定義乾淨的資料需要符合以下三項:

#### 每一個變數(variable)構成一欄(column)
#### 每一個觀察值(observation)構成一列(row)
#### 每一張表(table)只包含一種觀察單位(observational unit)
