```python
a , b , c = 1 ,2 ,3  # 多重指派  先打包  再解包

# a , b = 1, 2, 3 # too many values to unpack (expected 2)
# a , b , c = 1, 2  # not enough values to unpack (expected 3, got 2)
# a, b = 1 # cannot unpack non-iterable int object

a, b, c = [1,2,3]
print(a)    #1
a, b, c = (1,2,3)
print(a)    #1
```

```python
# a , b , c = 1 , ( 2 , 3 )  # not enough values to unpack (expected 3, got 2)
# '*' 在 '=' 右邊是解包, 在 '=' 左邊是打包


a , b , c = 1 , *( 2 , 3 ) # 強制解包

print(b)  #2
```
